```python
import pandas as pd 
import seaborn as sns 
import pyodbc
import numpy as np 
import matplotlib.pyplot as plt 
import sys 
sys.path.insert(1,'../py-scripts')
import matplotlib.ticker as mtick 

#Style Settings
sns.set(rc = {'axes.facecolor':'white'
              ,'figure.figsize':(12,8)
              ,'figure.facecolor':'white'
              ,'axes.edgecolor': '#C4C4BF'
              ,'lines.linewidth':4.5
              ,'axes.titlesize':26
              ,'axes.labelsize':16
              ,'axes.labelcolor':'black'
              ,'grid.color': '#EEEEED'
              ,'text.color':'black'
              ,'xtick.color':'black'
              ,'ytick.color':'black'
              ,'ytick.labelsize':14
              ,'xtick.labelsize':14
              ,'legend.fontsize':12
              ,'axes.linewidth':2
              ,'grid.linestyle': '--'},
            font = 'Segoe UI')
```

## Despined Graph

```python
x = np.arange(-5,5,0.1)
y = np.sin(x)

fig,ax = plt.subplots()
plt.title('Graph')
sns.lineplot(x = x, y = y)
plt.xlabel('X Values')
plt.ylabel('Y Values')
sns.despine(top = True, right = True, left = False, bottom = False)
ax.xaxis.grid(False)
```

## Percent Y Axis

```python
size = 1_000
x = np.sort(np.random.normal(size = size,scale = 4,loc = 10))
y = np.array(range(size))/float(size)


fig,ax = plt.subplots()
sns.lineplot(x = x , y = y)


plt.title('CDF Graph')
plt.xlabel('X Values')
plt.ylabel('Cumulative Density')
sns.despine(top = True, right = True, left = False, bottom = False)
ax.xaxis.grid(False)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1,decimals= 0))
```

## Shaded Distributions

```python
x1 = np.random.normal(loc = 0, scale = 1, size = 1000)
x2 = np.random.normal(loc = 0, scale = 1.4, size = 1000)

fig,ax = plt.subplots()
sns.despine(top = True, right = True, left = True, bottom = True)

sns.kdeplot(x1,color = "#33CCFF",label = 'x1')
sns.kdeplot(x2,color = '#002D5E',label = 'x2')


#Shading x1 Dist
x1_threshold = 1
x = ax.lines[0].get_xdata()
y = ax.lines[0].get_ydata()
ax.fill_between (x,0,y, where= x < x1_threshold, color = '#33CCFF',alpha = 0.1 )
ax.vlines(x1_threshold, 0 ,np.interp(x1_threshold,x,y), linestyle = '--',alpha = 0.5, color = '#33CCFF')


#Shading x2 Dist
x2_threshold = 0
x = ax.lines[1].get_xdata()
y = ax.lines[1].get_ydata()
ax.fill_between (x,0,y, where= x < x2_threshold, color = '#002D5E',alpha = 0.1 )
ax.vlines(x2_threshold, 0 ,np.interp(x2_threshold,x,y), linestyle = '--',alpha = 0.5, color = '#002D5E')

plt.title('Distribution Differences')
plt.xlabel('Scores')
plt.ylabel('')
plt.yticks([])
plt.show()
```
